```julia
# using...

using PyPlot
```

```julia
# sizes = NaN * zeros( length(labels) )
# explode = zeros( length(labels) )
# explode[ round( length( labels ) * rand(1) ) ] = 0.1

labels = ["Power","Transportation","Agriculture & land use","Buildings", "Manufacturing", "Consumer & service"]
colors = ["yellowgreen", "gold", "lightskyblue", "lightcoral", "chocolate", "ivory"]
explode = zeros( length(labels) ); explode[1] = .1
sizes = [2.0, 1.9, 1.6, 1.6, 1.2, 0.7]

pct = String[]
for i = 1:length(labels)
    push!(pct, string(sizes[i]))
end
total = sum(sizes)
function pp(x)
    return @sprintf("%1.1f\n", x*total/100)
end

fig = figure(figsize=(5,5)) # 10,10
p = pie(sizes,labels=labels,shadow=true,startangle=90,explode=explode,colors=colors,autopct=pp)

axis("equal")
title("ICT abatement potential by end-use sector\n")
PyPlot.savefig("pyplot.png")
```
